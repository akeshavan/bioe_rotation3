#Project Eurotrip

*The 10 subject calibration*

##Goal

We want to calculate the number of subjects needed to show ***equivalency*** of calibrated brain volume measurements between two different scanners. 

## Review

At each site, 50 age-matched controls from the same population are scanned. The scanner variability and subject variability are estimated from literature and from data in the lab. 

First, we define brain volume (BV) as a function of age for our healthy controls, with normally distributed variability:

###$BV = \alpha_a * A + N(B_0,\sigma_b)$

After being scanned at a site, the scanner scales the real brain volume and adds normally distributed scanner noise centered on an offset value:

At site 1:
###$\hat{BV_1} = \alpha_{s,1}*BV + N(\gamma_{s,1}, \sigma_{s,1})$

At site 2:
###$\hat{BV_2} = \alpha_{s,2}*BV + N(\gamma_{s,2}, \sigma_{s,2})$

## Calibration

To calibrate, we run an Ordinary Least Squares (OLS) model on each site, estimating the age coefficient and offset. 

###$\overline{BV_i} = \alpha_i*A+\gamma_i$

Ideally the fitted coefficient and intercept values for each site should be:

###$\alpha_i = \alpha_{s,i}*\alpha_a$
###$\gamma_i = \alpha_{s,i}*B_0+\gamma_{s,i}$

Our calibrated brain volumes for each subject (with age A) at each site are then:

###$Z_i(A) = \frac{\hat{BV_i(A)} - \overline{BV_i}(A)}{\sigma[\hat{BV_i}(A) - \overline{BV_i}(A)]} $

````python
def eq_test(pdiff,e):
    x = ro.FloatVector(pdiff)
    base = ro.packages.importr("base")
    res = eq.ptte_data(x,Epsilon=e)
    P = res.rx("Power")[0][0]
    T = res.rx("Tstat")[0][0]
    C = res.rx("cutoff")[0][0]
    return P,T,C
````

###Power Analysis for a Paired T Test

* Power is a function of $\alpha$, effect size, and number of people. 
* $\alpha$ = 0.05
* There are 10 controls
* All that is left is to define Effect Size, but first let us review the calibration model:

#### Calibration Model

We calibrate the scores from the two sides by doing the following:

####$Z_1 = \frac{BV_1 - \hat{BV_1}(A)}{\sigma(BV_1 - \hat{BV_1}(A))}$
####$Z_2 = \frac{BV_2 - \hat{BV_2}(A)}{\sigma(BV_2 - \hat{BV_2}(A))}$

Ideally the expectation of the pairwise difference in calibrated Brain volume should be very close to 0, but this is not the case:

####$|Z_{2,i}-Z_{1,i}| = d_i$ 

#### Effect Size

We define the effect size by **standardizing the mean difference** :

###$\tau = \frac{\mu(D)}{\sigma(D)}$
    
The code below calculates the effect size we see from calibration:    
```
    zdiff = np.abs(z_euro_1-z_euro_2)
    effect = np.mean(zdiff)/np.std(zdiff)
```
    
After running the model 1000 times, we eill the calculate the average effect size and then calculate the power, given that we have 10 subjects. Then, we will calculate the number of subjects needed for 80% power.

````python
P = []
Ps = []
Es = []
Ds = []
scan_rescan = range(1,10)
for i in scan_rescan:
    ps, pps, e, d = boot(10,n_scans=i,delta = 0)
    P.append(pps)
    Ps.append(ps)
    Es.append(e)
    Ds.append(d)
````
